```python
import PyPDF2
from pprint import pp
```

```python
def extract_text_page_by_page(pdf_path):
    with open(pdf_path, 'rb') as file:
        # Initialize PDF reader
        pdf_reader = PyPDF2.PdfReader(file, strict=True)
                
        # Iterate over all the pages and extract text
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            yield page.extract_text()

pdf_path = '../data/CS25Amendment27.pdf'

# Save the extracted text to text.txt
out_file = '../data/CS25Amendment27.txt'
with open(out_file, 'w') as file_obj:
    for page_text in extract_text_page_by_page(pdf_path):
        file_obj.write(page_text)

print(f"Text saved to {out_file}")
```
